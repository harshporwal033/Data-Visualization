```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline
```

```python
sns.set(style="whitegrid")
```

```python
df = sns.load_dataset("titanic")
```

```python
print("First 5 rows of dataset:")
print(df.head())
```

```python
df = df.dropna(subset=['age', 'fare', 'pclass', 'survived'])
```

```python
plt.figure(figsize=(10, 6))
age_counts = df['age'].value_counts().sort_index()

plt.plot(age_counts.index, age_counts.values, marker='o', linestyle='-', color='navy')
plt.title("Line Graph: Number of Passengers by Age")
plt.xlabel("Age")
plt.ylabel("Number of Passengers")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df['age'], bins=20, color='skyblue', edgecolor='black')
plt.title("Histogram: Distribution of Passenger Ages")
plt.xlabel("Age")
plt.ylabel("Number of Passengers")
plt.show()
```

```python
plt.figure(figsize=(8, 8))
survived_counts = df['survived'].value_counts()
labels = ['Did Not Survive', 'Survived']
colors = ['lightcoral', 'lightgreen']

plt.pie(survived_counts, labels=labels, autopct='%1.1f%%', colors=colors, startangle=140)
plt.title("Pie Chart: Survival Distribution")
plt.axis('equal')  # Equal aspect ratio ensures pie is a circle
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='pclass', y='fare', data=df, palette='Set3')
plt.title("Boxplot: Fare Distribution by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Fare")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(x='sex', y='survived', data=df, palette='pastel')
plt.title("Bar Plot: Survival Rate by Sex")
plt.xlabel("Sex")
plt.ylabel("Survival Rate")
plt.ylim(0, 1)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
corr = df[['survived', 'age', 'fare', 'pclass']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Heatmap: Correlation Matrix")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='pclass', hue='survived', data=df, palette='Set2')
plt.title("Count Plot: Survival by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Number of Passengers")
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.show()
```
